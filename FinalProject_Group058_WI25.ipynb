## Background and Prior Work


### Introduction

When considering the rates in which animals are admitted into shelters, a trend that often surfaces is the seasonality of intakes. Some months will see high influxes in admittance, while others seem to have very low rates of admittance. While this may partially be attributed to mating habits of animals occurring at different times of the year, we believe that the seasonality of animal shelter intakes may be slightly more complex, especially considering the wide range of breeding patterns of common pets like dogs and cats. 

In our analyses, we attempt to determine the structure of this seasonality, accounting for variance within intake patterns due to external factors, such as weather patterns and type of intake (owner-surrender vs. stray). With many years of data available, we aim to model seasonal trends within the different strata of intakes, hopefully identifying impactful trends that can be used to help manage shelter expectations for future years.  

### Prior Work

Reports showing animal shelter intake rates seem to display a similar pattern in overall admittance, with summer months showing an influx in intakes irrespective of location. Shelter Animals Count, a national animal shelter database site that publishes findings, has shown evidence of this seasonal trend having high variance based on year in a recent mid-year report.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) This report also stratifies by owner-surrender/stray and reported a moderate *increase* in owner-surrenders over time, mainly driven by cat surrenders. We aim to take a similar approach, focusing on longitudinal data analysis to see how these overall results generalize to the Austin Animal Shelter. However, we intend to focus more on the month-to-month differences in rates rather than year-to-year rates to better understand the scope of seasonality. 

Further studies have explored the variance in animal shelter intakes in longitudinal settings, with one such paper finding no significant change in intakes after accounting for rising human population rates.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This offers an interesting consideration in interpreting rates of intakes, especially if rates comove over time with population growth. While all of these studies focus on longitudinal analysis, they place heavy focus on yearly trends. A common throughline was mention of increasing yearly intake rates<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3), though rarely do these studies focus on the smaller-scale changes in time, which we intend to do. We believe that analysis of intake data within this lens will allow shelters to become better prepared in the short-term, potentially altering resource allocation month-by-month to match seasonal findings.

# Hypothesis


We hypothesize that the rate of animal intakes at the Austin Animal Center fluctuates seasonally, with intake rates peaking in the summer months and declining in the winter. Specifically, we predict that stray intakes will be higher in warmer months due to increased outdoor activity, higher birth rates for animals, and a greater likelihood of pet loss or abandonment. In contrast, we expect owner surrenders to peak after the holiday season, driven by financial strain, housing changes, or the realization that families cannot keep pets received as gifts.

### Rationale
Stray Intakes: Warmer weather encourages more outdoor activity, increasing the chances of pets getting lost or abandoned. Additionally, stray populations may grow during these months due to higher birth rates, particularly for feral cats and unaltered pets.

Owner Surrenders: Post-holiday periods often see a rise in pet relinquishments as new pet owners face unexpected costs, lifestyle changes, or housing restrictions that make keeping a pet unfeasible.

These seasonal trends align with historical patterns observed in animal shelters, where external factors such as weather conditions and social behaviors impact intake rates. Understanding these trends can help shelters allocate resources more effectively and plan for intake fluctuations throughout the year.

```python
#Importing all necessary packages for this project
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

df = pd.read_csv('intakes.csv')
df.head
```

Looking at our dataframe, we can remove any unnecessary columns that are not related to our question such as name, or animal ID. Although animal IDs can help us distinguish unique animals, similar to a primary key in SQL, it is not needed in regards to our research question or project

```python
df = df.drop(columns=['Name', 'Animal ID', 'Color', 'Breed'])
```

Now let's check for any null or missing values in our dataset. If there are, we may have to impute values based on metrics.

```python
print(df.isnull().any())
print("Number of NaN values in Intakes.csv: " + str(df.isnull().sum().sum()))
```

```python
df = df.dropna(subset=['Sex upon Intake', 'Age upon Intake'])
print("Number of NaN values in Intakes.csv: " + str(df.isnull().sum().sum()))
```

After making sure there are no missing values, we can check the shape of our dataframe once again. We should have 3 less columns because they were already dropped, and we also should have two less rows for the null values. 

```python
print(df.shape)
print(df.describe())
```

Looking at the count of animals types, the three main animal types were cat, dog, and other. The count of livestock and birds were too low for us to stratify for significant analysis, so we decided to drop all rows that include 'bird' or 'livestock' in the "Animal Type" Column

```python
print(df['Animal Type'].value_counts())
df = df[~df['Animal Type'].isin(['Livestock', 'Bird'])]
print(df['Animal Type'].value_counts())
```

When analyzing the 'Age upon Intake' variable, we can see that there is a huge variety of possible values - and nothing is standardized (with days, weeks, months, and years appearing). More importantly, we see negative values of age which we decided to omit from our future analyses.

```python
print(df['Age upon Intake'].unique())
```

First, we standardized the age of all intakes to months and then dropped those with negative initial age values. We got rid of 9 observations with negative ages.

```python
#Converting Intake Age to Months
def standardize_time(time):
    number, unit = time.split()
    number = int(number)

    if "year" in unit:
        return number 
    elif "month" in unit or "months" in unit:
        return number / 12
    elif "week" in unit or "weeks" in unit:
        return number / 52  
    elif "day" in unit:
        return number / 365
    else:
        return None  

df['Age upon Intake'] = df['Age upon Intake'].apply(standardize_time)

obs_before = len(df)
df = df[df['Age upon Intake'] >= 0]
obs_after = len(df)

print("\033[1mObservations dropped:\033[0m " + str(obs_before - obs_after))
```

Second to lastly, we have an address/ found location column, but we will not be needing the specific address of the where the animal was found, instead a city will suffice. Using the .extract function we can get the last word in the string and also drop the (TX) at the end of the address as well.

```python
df['Found Location'] = df['Found Location'].str.extract(r'(\b[A-Za-z]+)(?:\s*\(TX\))?$')
print(df.head)
print(df.shape)
```

Before we start plotting and regressing data, we should make sure that time is in the right format / type. This is especially useful when plotting date on the x-axis, as it allows longitudinal analyses. when To do this we use pd.datetime and then check the ranges to see if it worked (it did)!

```python
print("Fixing Date Types\n")
#Looking at data types
print("\033[1mData Types:\033[0m \n" + str(df.dtypes))

#Converting DateTime into correct type
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%m/%d/%Y %I:%M:%S %p')
df['MonthYear'] = pd.to_datetime(df['MonthYear'], format='%B %Y')

print("\n\033[1mNew Type of 'DateTime':\033[0m \n" + str(df['DateTime'].dtypes))
print("\n\033[1mNew Type of 'MonthYear':\033[0m \n" + str(df['MonthYear'].dtypes))

#Looking at the range of years included
print("\n\033[1mRange of years (Full):\033[0m\n" + str(df['DateTime'].min().date()) + " to " + str(df['DateTime'].max().date()))
print("\n\033[1mRange of years (M/Y):\033[0m\n" + str(df['MonthYear'].min().date()) + " to " + str(df['MonthYear'].max().date()))
```

Then we define the function assign_season, which categorizes each month into one of four seasons: Winter (Decemberâ€“February), Spring (Marchâ€“May), Summer (Juneâ€“August), and Fall (Septemberâ€“November). This function is applied to the dataset to create a new Season column.

```python
df_merged['month'] = pd.to_numeric(df_merged['month'], errors='coerce')
df_merged['month'] = df_merged['month'].fillna(1).astype(int)

def assign_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
    
df_merged['Season'] = df_merged['month'].apply(assign_season)
```
